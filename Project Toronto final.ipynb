```python
import pandas as pd
import numpy as np
```

### Reading the file ###

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
df[0]
```

```python
df[0].head()
```

### Dropping rows in neighborhood with NaN values and resetting index ###  

```python
df[0].dropna(subset=["Neighborhood"], axis=0, inplace = True)
df[0].reset_index(drop=True, inplace = True)
df[0]
```

### Obtain the number of rows and columns ###

```python
df[0].shape
```

### Read the location file and drop the postal code column ###

```python
Coordinate=pd.read_csv('https://cocl.us/Geospatial_data')
Coordinate.head()
Coordinate.drop("Postal Code", axis=1, inplace = True)
Coordinate
```

### Merge the two tables ###

```python
df1 = pd.concat([df[0], Coordinate], axis=1)
df1
```

```python
from geopy.geocoders import Nominatim
address = 'Toronto, TO'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium
# create map of toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df1['Latitude'], df1['Longitude'], df1['Borough'], df1['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
scar_data = df1[df1['Borough'] == 'Downtown Toronto'].reset_index(drop=True)
scar_data.head()
```

```python
address = 'Downtown Toronto, TO'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geographical coordinates of Downtown Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Downtown Toronto using latitude and longitude values
map_scar = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(scar_data['Latitude'], scar_data['Longitude'], scar_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_scar)  
    
map_scar
```
